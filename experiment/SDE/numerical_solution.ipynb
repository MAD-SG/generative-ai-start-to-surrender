```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.integrate import solve_ivp

# Define the SDE: dX = mu(X) dt + sigma(X) dW
def mu(x):
    return -x  # Ornstein-Uhlenbeck drift term

def sigma(x):
    return 1.0  # Constant diffusion term

# Simulate SDE using Euler-Maruyama
np.random.seed(42)
T = 5.0  # Total time
dt = 0.001  # Smaller time step for accuracy
N = int(T / dt)  # Number of steps
M = 10000  # Number of particles

time_points = [0.5, 1.0, 2.0, 5.0]  # Time points for comparison
X = np.zeros((M, N))
X[:, 0] = np.random.normal(0, 1, M)  # Initial condition

for i in range(1, N):
    dW = np.sqrt(dt) * np.random.randn(M)  # Wiener process increment
    X[:, i] = X[:, i - 1] + mu(X[:, i - 1]) * dt + sigma(X[:, i - 1]) * dW

# Define the Fokker-Planck equation
def fokker_planck(t, p, x_grid):
    dpdt = np.zeros_like(p)
    dx = x_grid[1] - x_grid[0]

    for i in range(1, len(x_grid) - 1):
        diffusion_term = (sigma(x_grid[i])**2 * (p[i+1] - 2*p[i] + p[i-1]) / dx**2)
        drift_term = -mu(x_grid[i]) * (p[i+1] - p[i-1]) / (2*dx)
        dpdt[i] = diffusion_term + drift_term

    # Zero flux boundary conditions
    dpdt[0] = dpdt[1]
    dpdt[-1] = dpdt[-2]

    return dpdt

# Solve the Fokker-Planck equation at multiple time points
x_grid = np.linspace(-4, 4, 500)
p0 = norm.pdf(x_grid, 0, 1)  # Initial condition
solution = solve_ivp(fokker_planck, [0, max(time_points)], p0, t_eval=time_points, args=(x_grid,))

# Normalize FPE solutions
p_fpe_solutions = []
for i in range(len(time_points)):
    p_fpe = solution.y[:, i]
    p_fpe /= np.trapz(p_fpe, x_grid)
    p_fpe_solutions.append(p_fpe)

# Plot results at different time points
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, t in enumerate(time_points):
    # Get SDE results at closest time index
    time_index = int(t / dt)
    hist_bins = np.linspace(-4, 4, 50)
    hist_sde, _ = np.histogram(X[:, time_index], bins=hist_bins, density=True)

    axes[i].hist(X[:, time_index], bins=hist_bins, density=True, alpha=0.5, label=f'SDE Simulation (t={t})')
    axes[i].plot(x_grid, p_fpe_solutions[i], label=f'FPE Solution (t={t})', color='r')
    axes[i].set_xlabel('X')
    axes[i].set_ylabel('Probability Density')
    axes[i].legend()
    axes[i].set_title(f'Distribution at t={t}')

plt.tight_layout()
plt.show()
```

```python
# prerequisites
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.distributions.normal import Normal
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import numpy as np

class funnel():
    def __init__(self, y_std=1):
        self.y_dist = Normal(0, y_std)

    def log_prob_normal(self, mu, scale, x):
        cov = scale*scale
        a = -0.5 * (np.log(2 * np.pi) + torch.log(cov))
        b =  -0.5 * (x - mu) * (x - mu) / cov
        return a + b

    def log_prob(self, data):
        x = data[:, 0]
        y = data[:, 1]
        y_log_pdf = self.y_dist.log_prob(y)
        x_log_pdf = self.log_prob_normal(0.0, self.x_std(y), x)
        return x_log_pdf + y_log_pdf

    def x_std(self, y):
        return torch.exp(y) / 10

    def sample(self, n_samples):
        y = self.y_dist.sample([n_samples])
        x = torch.randn_like(y) * self.x_std(y)
        return torch.vstack([x, y]).T
funnel_dist = funnel()
```

```python
def plt_potential_func(log_prob, ax, num_grid=500, xs=[-5, 5], ys=[-5, 5], cmap=None):

    xlist = torch.linspace(xs[0], xs[1], num_grid)
    ylist = torch.linspace(ys[0], ys[1], num_grid)
    xx, yy = np.meshgrid(xlist, ylist)
    z = np.hstack([xx.reshape(-1, 1), yy.reshape(-1, 1)])

    z = torch.Tensor(z)
    u = log_prob(z).cpu().numpy()
    p = u.reshape(num_grid, num_grid)
    return ax.pcolor(xx, yy, p, cmap=cmap, vmin=-7.5, vmax=1.0)
fig, ax = plt.subplots(figsize=(4, 4), frameon=False)
im = plt_potential_func(funnel_dist.log_prob, ax)
ax.axis('equal')
ax.axis("off")
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.margins(0,0)
display(fig)
plt.close()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output, display
from matplotlib.animation import FuncAnimation, PillowWriter
from tqdm import tqdm

# -------------------------------
# Define the gradient calculation using PyTorch autograd
# -------------------------------
def grad_logp(x, logp):
    # Compute the gradient of logp(x) with respect to x.
    # grad_outputs=torch.ones_like(logp(x)) ensures proper backpropagation.
    return torch.autograd.grad(logp(x), x, grad_outputs=torch.ones_like(logp(x)),
                               create_graph=False)[0]

# -------------------------------
# Initialize particles
# -------------------------------
# 1024 particles in 2D; shifted so that the initial distribution is centered at [0, 2]
xs_lang = torch.randn([1024, 2]) * 0.5 + torch.tensor([.0, 2.0])

# -------------------------------
# Setup the figure for visualization
# -------------------------------
fig, ax = plt.subplots(figsize=(4, 4))
scatter_plot = ax.scatter(xs_lang[:, 0].detach().numpy(), xs_lang[:, 1].detach().numpy(),
                          s=5, alpha=0.2, color='red')
time_text = ax.text(1.0, -3.5, '$t=0$', fontsize=14)
ax.axis('equal')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.axis("off")  # remove axes to speed up drawing
fig.tight_layout()

# Display the initial figure (useful in Jupyter notebooks)
clear_output(wait=True)
display(fig)

# Enable gradient tracking on xs_lang
xs_lang.requires_grad = True
flows = [xs_lang.detach()]

# -------------------------------
# Langevin dynamics parameters and loop
# -------------------------------
n_iters = 1500      # number of iterations
lr = 1e-5           # initial learning rate (for monitoring; not used in update)
decay_rate = 0.999  # learning rate decay

# Unadjusted Langevin dynamics loop:
for i in range(n_iters + 1):
    # Update step: current state + drift (step size * gradient) + diffusion noise
    xs_lang = xs_lang + 0.001 * grad_logp(xs_lang, funnel_dist.log_prob) \
              + np.sqrt(0.001 * 2) * torch.randn_like(xs_lang)

    # Decay learning rate (for monitoring purposes)
    lr = lr * decay_rate

    # Save the current state for later animation
    flows.append(xs_lang.detach())

    # Update the scatter plot and time text
    scatter_plot.set_offsets(xs_lang.detach().numpy())
    time_text.set_text(f'$t={i}/{n_iters}$')

    # Update the display (suitable for Jupyter notebooks)
    clear_output(wait=True)
    display(fig)
    print('[%d/%d]: lr, %.7f' % (i, n_iters, lr))

# Close the plot once iterations are complete
plt.close()

# -------------------------------
# Create and save the animation as a GIF
# -------------------------------
def update(frame):
    scatter_plot.set_offsets(flows[frame].numpy())
    time_text.set_text(f'$t={frame}/{n_iters}$')

# Create the animation using the saved flows
ani = FuncAnimation(fig=fig, func=update, frames=tqdm(range(len(flows))), interval=1)
writer = PillowWriter(fps=40)
ani.save("funnel_langevin.gif", writer=writer)
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from tqdm import trange

# -------------------------------
#  Funnel  log 
# z = (x, v) v ~ N(0, 3^2) x|v ~ N(0, exp(v))
# -------------------------------
def funnel_log_prob(z):
    # z: [N, 2]  x v
    x = z[:, 0]
    v = z[:, 1]
    lp_v = -0.5 * (v / 3)**2 - torch.log(3 * torch.sqrt(torch.tensor(2 * np.pi)))
    lp_x_given_v = -0.5 * torch.log(torch.tensor(2 * np.pi)) - 0.5 * v - 0.5 * (x**2) / torch.exp(v)
    return lp_v + lp_x_given_v

# -------------------------------
#  autograd  Funnel  z log p_funnel(z)
# -------------------------------
def grad_logp_funnel(z):
    #  z  requires_grad=True
    z = z.clone().detach().requires_grad_(True)
    lp = funnel_log_prob(z)
    grad = torch.autograd.grad(lp.sum(), z)[0]
    return grad

# -------------------------------
#  ODE  drift 
#  ODE 
#   dz/dt = +0.5 * [z log p_funnel(z) + z]
# -------------------------------
def ode_drift_reverse(z, t):
    #  drift  t
    grad_logp = grad_logp_funnel(z)
    return 0.5 * (grad_logp + z)

# -------------------------------
# RK4 
# -------------------------------
def rk4_step(z, t, dt, drift_fn):
    k1 = drift_fn(z, t)
    k2 = drift_fn(z + 0.5 * dt * k1, t + 0.5 * dt)
    k3 = drift_fn(z + 0.5 * dt * k2, t + 0.5 * dt)
    k4 = drift_fn(z + dt * k3, t + dt)
    return z + (dt / 6.0) * (k1 + 2*k2 + 2*k3 + k4)

# -------------------------------
#  ODE  t=0  t=1 t  0  1 t 
#  t=0  t=1  Funnel 
# 
# -------------------------------
def solve_reverse_ode(zT, n_steps=500):
    dt = 1.0 / n_steps
    t = 0.0
    z = zT
    trajectory = [zT.detach().cpu().numpy()]
    for i in range(n_steps):
        z = rk4_step(z, t, dt, ode_drift_reverse)
        t += dt
        trajectory.append(z.detach().cpu().numpy())
    return z, trajectory

# -------------------------------
# 
# -------------------------------
# 
N = 1024
# t=1 
zT = torch.randn(N, 2)

#  ODE Funnel 
z0, traj = solve_reverse_ode(zT, n_steps=1500)

# 
fig, ax = plt.subplots(figsize=(5,5))
scat = ax.scatter(traj[0][:,0], traj[0][:,1], s=5, alpha=0.5, color='red')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_title("ODE ")
ax.set_xlabel("x")
ax.set_ylabel("v")

# 
def update(frame):
    data = traj[frame]
    scat.set_offsets(data)
    ax.set_title(f"ODE sampling  [{frame}/{len(traj)-1}]")
    return scat,

# 

#  GIF 
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from tqdm import trange

# ===============================
# 1.  Funnel  score 
# ===============================
def funnel_log_prob(z):
    """
     Funnel 
      v ~ N(0, 3^2)
      x|v ~ N(0, exp(v))
     z  shape  [N, 2] x v
    """
    x_val = z[:, 0]
    v_val = z[:, 1]
    lp_v = -0.5 * (v_val / 3)**2 - torch.log(3 * torch.sqrt(torch.tensor(2 * np.pi)))
    lp_x_given_v = -0.5 * torch.log(torch.tensor(2 * np.pi)) - 0.5 * v_val - 0.5 * (x_val**2) / torch.exp(v_val)
    return lp_v + lp_x_given_v

def score_function(x, t):
    """
     t  score:
      s(x,t)  (1-t)*_x log p_funnel(x) + t*(-x)
     t=0  Funnel  scoret=1  score-x
    """
    #  score
    x_in = x.clone().detach().requires_grad_(True)
    lp = funnel_log_prob(x_in)
    score_target = torch.autograd.grad(lp.sum(), x_in)[0]
    return (1 - t) * score_target + t * (-x_in)

# ===============================
# 2. 
# ===============================
def beta_t(t, beta_min=0.1, beta_max=20.0):
    """: beta(t) = beta_min + t*(beta_max-beta_min)"""
    return beta_min + t * (beta_max - beta_min)

# ===============================
# 3.  DDPM  SDE 
# ===============================
def ddpm_reverse_sde(num_steps=1000, batch_size=1024, device='cpu'):
    """
     EulerMaruyama  SDE 
      dx = [ -0.5*beta(t)*x - beta(t)*s(x,t) ] dt + sqrt(beta(t))*dW,
     dt  t=1  t=0
    
      num_steps: 
      batch_size: 
      device: 
    
      
    """
    dt = -1.0 / num_steps  # dt 10
    # t=1 
    x = torch.randn(batch_size, 2, device=device)
    trajectory = [x.clone().detach().cpu().numpy()]
    t = 1.0
    for i in trange(num_steps):
        beta = beta_t(t)
        #  SDE  drift  -0.5*beta*x - beta*s(x,t)
        drift = -0.5 * beta * x - beta * score_function(x, t)
        # diffusion  sqrt(beta(t)*|dt|)
        diffusion = torch.sqrt(torch.tensor(beta * abs(dt), dtype=x.dtype, device=x.device)) * torch.randn_like(x)
        # EulerMaruyama 
        x = x + drift * dt + diffusion
        trajectory.append(x.clone().detach().cpu().numpy())
        t = t + dt  # dt t 
    return x, trajectory

# ===============================
# 4. 
# ===============================
device = 'cpu'
num_steps = 1000
batch_size = 1024

#  SDE  Funnel 
final_sample, traj = ddpm_reverse_sde(num_steps=num_steps, batch_size=batch_size, device=device)

# t=1, t=0, Funnel
plt.grid('off')
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(traj[0][:, 0], traj[0][:, 1], s=5, alpha=0.5)
plt.title("t = 1 (initial sampling)")
plt.subplot(1, 2, 2)
plt.scatter(traj[-1][:, 0], traj[-1][:, 1], s=5, alpha=0.5, color='red')
plt.title("t = 0 (inverse SDE final Funnel sampling)")
plt.show()

# 
fig, ax = plt.subplots(figsize=(5, 5))
scat = ax.scatter(traj[0][:, 0], traj[0][:, 1], s=5, alpha=0.5, color='red')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_title("DDPM inverse SDE sampling")
ax.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)

def update(frame):
    data = traj[frame]
    scat.set_offsets(data)
    current_t = 1 - frame / num_steps  # t  1  0
    ax.set_title(f"t = {current_t:.3f}")
    return scat,

ani = FuncAnimation(fig, update, frames=len(traj), interval=20, blit=True)
writer = PillowWriter(fps=30)
ani.save("ddpm_reverse_sde_sampling.gif", writer=writer)
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from tqdm import trange

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# ===============================
# 1. Funnel Distribution Sampling
# ===============================
def sample_funnel(batch_size):
    """
    Sample from the Funnel distribution.
    For each sample:
      - v ~ N(0, 3^2)   (standard deviation 3)
      - x | v ~ N(0, exp(v))
    Returns a tensor of shape [batch_size, 2] where
      sample[:, 0] is x and sample[:, 1] is v.
    """
    v = torch.randn(batch_size, 1, device=device) * 3.0  # v ~ N(0,9)
    std_x = torch.exp(v / 2)  # standard deviation for x is exp(v/2)
    x = torch.randn(batch_size, 1, device=device) * std_x
    return torch.cat([x, v], dim=1)

# ===============================
# 2. Continuous Forward Diffusion Process
# ===============================
def forward_diffusion_cont(x0, t, beta_start=1e-4, beta_end=0.02):
    """
    Compute the forward diffusion process for continuous time.

    Args:
      x0: Original samples, tensor of shape [batch_size, d]
      t:  Continuous time, tensor of shape [batch_size, 1] with values in [0,1]
      beta_start, beta_end: scalars defining the continuous noise schedule.

    We define:
        (t) = exp( - (beta_start * t + 0.5*(beta_end - beta_start) * t^2) )
    Then:
        x_t = sqrt((t)) * x0 + sqrt(1 - (t)) * ,   ~ N(0,I)
    """
    alpha_t = torch.exp(-(beta_start * t + 0.5 * (beta_end - beta_start) * t * t))
    noise = torch.randn_like(x0)
    sqrt_alpha_t = torch.sqrt(alpha_t)
    sqrt_one_minus_alpha_t = torch.sqrt(1.0 - alpha_t)
    x_t = sqrt_alpha_t * x0 + sqrt_one_minus_alpha_t * noise
    return x_t, noise, alpha_t

# ===============================
# 3. Define the Noise Prediction Network (Score Network)
# ===============================
class ScoreNetCont(nn.Module):
    def __init__(self, input_dim=2, hidden_dim=128, output_dim=2):
        """
        A simple MLP that takes x_t (dim 2) and a continuous time t (dim 1)
        concatenated (total dim=3) and outputs the predicted noise (dim 2).
        """
        super(ScoreNetCont, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim + 1, hidden_dim),
            nn.SiLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.SiLU(),
            nn.Linear(hidden_dim, output_dim)
        )

    def forward(self, x, t):
        # x is [batch_size, 2] and t is [batch_size, 1]
        inp = torch.cat([x, t], dim=1)
        return self.model(inp)

# ===============================
# 4. Continuous DDPM Training Function
# ===============================
def train_ddpm_continuous(model, beta_start=0.0, beta_end=100,
                          batch_size=256, num_steps=10000, lr=1e-3):
    """
    Train a DDPM noise-prediction model in continuous time.

    Forward process:
        x_t = sqrt((t)) * x0 + sqrt(1-(t)) * ,
    where (t) = exp(-[beta_start*t + 0.5*(beta_end-beta_start)*t^2]).

    Loss: MSE between the predicted noise and the true noise.
    """
    optimizer = optim.Adam(model.parameters(), lr=lr)
    mse_loss = nn.MSELoss()
    loss_history = []

    for step in trange(1, num_steps+1):
        # 1) Sample x0 from funnel distribution
        x0 = sample_funnel(batch_size)  # shape: [batch_size, 2]

        # 2) Sample continuous time t uniformly in [0,1]
        t = torch.rand(batch_size, 1, device=device)

        # 3) Forward diffusion: get x_t, true noise, and (t)
        x_t, true_eps, alpha_t = forward_diffusion_cont(x0, t, beta_start, beta_end)

        # 4) Predict noise with the network given x_t and t
        pred_eps = model(x_t, t)

        # 5) Compute MSE loss between predicted and true noise
        loss = mse_loss(pred_eps, true_eps)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        loss_history.append(loss.item())

        if step % 500 == 0:
            print(f"Step {step}/{num_steps}, Loss: {loss.item():.6f}")

    return loss_history

# ===============================
# 5. Train the Model in Continuous Time
# ===============================
if __name__ == "__main__":
    # Instantiate the continuous time DDPM model
    model_cont = ScoreNetCont(input_dim=2, hidden_dim=256, output_dim=2).to(device)

    loss_hist = train_ddpm_continuous(model_cont, beta_start=0.0, beta_end=30.0,
                                      batch_size=128, num_steps=20000, lr=1e-5)

    # Plot training loss
    plt.plot(loss_hist)
    plt.xlabel("Training Step")
    plt.ylabel("MSE Loss")
    plt.title("Continuous DDPM Training Loss on Funnel Distribution")
    plt.show()
```

```python
import numpy as np
import torch
import matplotlib.pyplot as plt

def plot_gradient_field(model, t_value, grid_range=(-4, 4), grid_points=20,
                        beta_start=1e-4, beta_end=0.02):
    """
    Plots the learned gradient field (score function) from the trained DDPM model.

    Args:
        model: Trained continuous-time DDPM noise predictor (e.g., model_cont).
        t_value: The continuous time value in [0, 1] at which to evaluate the score.
        grid_range: Tuple (min, max) for both dimensions.
        grid_points: Number of points per axis.
        beta_start, beta_end: Noise schedule parameters.
    """
    device = next(model.parameters()).device
    # Create a grid of (x, v) points.
    x_lin = np.linspace(grid_range[0], grid_range[1], grid_points)
    v_lin = np.linspace(grid_range[0], grid_range[1], grid_points)
    X, V = np.meshgrid(x_lin, v_lin)

    # Flatten grid and create tensor of shape [N, 2]
    grid_np = np.stack([X.ravel(), V.ravel()], axis=1)
    grid_tensor = torch.tensor(grid_np, device=device, dtype=torch.float32)

    # Create a time tensor for all grid points (shape [N, 1])
    t_tensor = torch.full((grid_tensor.shape[0], 1), t_value, device=device)

    # Get the noise prediction from the model and compute the approximate score.
    with torch.no_grad():
        pred_eps = model(grid_tensor, t_tensor)  # shape [N, 2]
        # Compute alpha(t) using your continuous formulation.
        alpha_val = alpha_t_fn(t_tensor, beta_start, beta_end)  # shape [N, 1]
        # The approximate score function:
        # _x log p_t(x)  -pred_eps / sqrt(1 - alpha(t))
        score = -pred_eps / torch.sqrt(1.0 - alpha_val)

    # Convert score to NumPy array and reshape to grid for plotting.
    score = score.cpu().numpy()
    U = score[:, 0].reshape(X.shape)
    W = score[:, 1].reshape(V.shape)

    # Plot the quiver field.
    plt.figure(figsize=(8, 8))
    plt.quiver(X, V, U, W, color='blue', scale=1, scale_units='xy')
    plt.title(f"Learned Gradient Field at t = {t_value}")
    plt.xlabel("x")
    plt.ylabel("v")
    plt.grid(True)
    plt.show()

```

```python
import torch

def beta_t_fn(t, beta_start=1e-4, beta_end=20.0):
    """
    Continuous beta(t) in [0,1].
    Here we do a simple linear schedule:
        beta(t) = beta_start + t * (beta_end - beta_start)
    """
    return beta_start + t * (beta_end - beta_start)

def alpha_t_fn(t, beta_start=1e-4, beta_end=0.02):
    """
    alpha(t) = exp( - [beta_start * t + 0.5*(beta_end-beta_start)*t^2] ).
    This matches the integral of beta(t) from 0 to t.
    """
    return torch.exp(-(beta_start * t + 0.5 * (beta_end - beta_start) * t * t))
def get_score(model, x, t, beta_start=1e-4, beta_end=0.02):
    """
    Approximate _{x_t} log p_t(x_t) using the trained model.

    x: [batch_size, d] current sample
    t: float in [0,1] (the continuous time)
    """
    # Create a time tensor for each sample in the batch
    batch_size = x.shape[0]
    t_tensor = torch.full((batch_size, 1), t, device=x.device)

    with torch.no_grad():
        # Model predicts the noise (epsilon)
        pred_eps = model(x, t_tensor)  # shape [batch_size, d]

    # Compute alpha(t)
    alpha_val = alpha_t_fn(t_tensor, beta_start, beta_end)  # shape [batch_size, 1]
    # Score = - pred_eps / sqrt(1 - alpha(t))
    score = -pred_eps / torch.sqrt(1.0 - alpha_val)
    return score
```

```python
def sample_sde(model, num_steps=1000, batch_size=1024,
               beta_start=1e-4, beta_end=0.02):
    device = next(model.parameters()).device

    # Start from Gaussian noise at t=1
    x = torch.randn(batch_size, 2, device=device)

    dt = 1.0 / num_steps
    trajectory = []

    t = 1.0
    for i in range(num_steps + 1):
        trajectory.append(x.clone().detach().cpu().numpy())

        if i < num_steps:
            # b_val is a float
            b_val = beta_t_fn(t, beta_start, beta_end)
            # Convert to PyTorch tensor
            b_t = torch.tensor(b_val, device=device, dtype=x.dtype)

            # Score
            sc = get_score(model, x, t, beta_start, beta_end)

            # Drift: -0.5 * b_t * x - b_t * sc
            drift = -0.5 * b_t * x - b_t * sc

            # Diffusion: sqrt(b_t * dt) * Normal(0, I)
            noise = torch.randn_like(x)
            diffusion = torch.sqrt(b_t * dt) * noise

            # Update
            x = x + drift * dt + diffusion

            t = t - dt

    return trajectory
```

```python
def sample_ode(model, num_steps=1000, batch_size=1024,
               beta_start=1e-4, beta_end=0.02):
    """
    Deterministic probability flow ODE sampling from t=1 to t=0 in 'num_steps' steps.
    Returns a list of [batch_size, 2] arrays representing the trajectory.
    """
    device = next(model.parameters()).device

    # Start from Gaussian noise at t=1
    x = torch.randn(batch_size, 2, device=device)

    dt = 1.0 / num_steps
    trajectory = []

    t = 1.0
    for i in range(num_steps + 1):
        trajectory.append(x.clone().detach().cpu().numpy())

        if i < num_steps:
            # Current beta(t)
            b_t = beta_t_fn(t, beta_start, beta_end)

            # Score
            sc = get_score(model, x, t, beta_start, beta_end)

            # Drift: -0.5 * b(t) * [ x + score ]
            drift = -0.5 * b_t * (x + sc)

            # Euler update
            x = x + drift * dt

            t = t - dt

    return trajectory
```

```python
class funnel():
    def __init__(self, y_std=1.0):
        self.y_dist = Normal(0.0, y_std)

    def log_prob_normal(self, mu, scale, x):
        cov = scale * scale
        a = -0.5 * (np.log(2 * np.pi) + torch.log(cov))
        b = -0.5 * (x - mu)*(x - mu) / cov
        return a + b

    def x_std(self, y):
        """funnel  x : exp(y)/10"""
        return torch.exp(y) / 10

    def log_prob(self, data):
        """
        data: shape [batch_size, 2],  data[:,0] = x, data[:,1] = y
         funnel  log_prob
        """
        x = data[:, 0]
        y = data[:, 1]
        # y_log_pdf
        y_log_pdf = self.y_dist.log_prob(y)
        # x_log_pdf
        x_log_pdf = self.log_prob_normal(0.0, self.x_std(y), x)
        return x_log_pdf + y_log_pdf

    def sample(self, n_samples):
        """
         n_samples  (x, y)
        y ~ N(0,1),
        x ~ N(0, (exp(y)/10)^2)
        """
        y = self.y_dist.sample((n_samples,))
        x = torch.randn_like(y) * self.x_std(y)
        return torch.stack([x, y], dim=1)
```

```python
class VESDE:
    def __init__(self, sigma_min=0.01, sigma_max=50.0, T=1.0):
        self.sigma_min = sigma_min
        self.sigma_max = sigma_max
        self.T = T

    def sigma(self, t):
        """ schedule [0,1] """
        return self.sigma_min * (self.sigma_max / self.sigma_min) ** t

    def perturb_x(self, x0, t):
        """
         t x0 
        x_t ~ N(x0, sigma(t)^2 I)
         (x_t, t) DSM 
        """
        std = self.sigma(t)
        noise = torch.randn_like(x0)
        return x0 + std * noise
```

```python
class ScoreNet(nn.Module):
    def __init__(self, dim=2, hidden_dim=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(dim + 1, hidden_dim),
            nn.SiLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.SiLU(),
            nn.Linear(hidden_dim, dim)
        )

    def forward(self, x, t):
        """
        x: shape [batch_size, 2]
        t: shape [batch_size, 1]
        : shape [batch_size, 2]  (score estimate)
        """
        inp = torch.cat([x, t], dim=1)  # 
        return self.net(inp)
```

```python
def dsm_loss(score_net, sde, x0):
    """
    x0: [batch_size, 2],  funnel 
    """
    batch_size = x0.size(0)
    # 1)  t ~ Uniform(0,1)
    t = torch.rand(batch_size, 1, device=x0.device)

    # 2)  x_t = x0 + sigma(t)*noise
    x_t = sde.perturb_x(x0, t)

    # 3)  score ( x_t  log)
    sigma_t = sde.sigma(t)
    target_score = -(x_t - x0) / (sigma_t**2)  # shape [batch_size, 2]

    # 4) 
    pred_score = score_net(x_t, t)

    # 5) MSE
    loss = 0.5 * (pred_score - target_score).pow(2).sum(dim=1).mean()
    return loss
```

```python
def train_score_network(score_net, sde, funnel_dist,
                        epochs=2000, batch_size=256, lr=1e-3, device='cpu'):
    optimizer = optim.Adam(score_net.parameters(), lr=lr)

    #  DataLoader
    data_all = funnel_dist.sample(20000).to(device)

    for epoch in range(epochs):
        idx = torch.randint(0, data_all.shape[0], (batch_size,), device=device)
        x0 = data_all[idx]  # shape [batch_size, 2]

        loss = dsm_loss(score_net, sde, x0)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if (epoch+1) % 500 == 0:
            print(f"Epoch {epoch+1}/{epochs}, loss = {loss.item():.6f}")
```

```python
@torch.no_grad()
def sample_reverse_sde(score_net, sde, num_samples=1000, num_steps=500, device='cpu'):
    """
     Euler-Maruyama  SDE:
    1)  N(0, sigma_max^2 I) 
    2)  T->0 
    3) : x_{t+dt} = x_t + drift * dt + diffusion * dW
    """
    # : 
    x = torch.randn(num_samples, 2, device=device) * sde.sigma_max

    t_space = torch.linspace(sde.T, 0, num_steps, device=device)
    dt = -sde.T / num_steps  # 

    for i in range(num_steps-1):
        t_cur = t_space[i].view(-1,1)  # shape [1,1]

        #  score
        score = score_net(x, t_cur.repeat(num_samples,1))

        sigma_t = sde.sigma(t_cur)
        # drift = - sigma(t)^2 * score
        drift = -(sigma_t**2) * score

        # diffusion = sigma(t)* sqrt(-dt)
        dw = torch.randn_like(x) * torch.sqrt(torch.tensor(-dt, device=device)) * sigma_t

        x = x + drift * dt + dw

    return x
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print("Using device:", device)

# 1) funnel
funnel_dist = funnel(y_std=1.0)

# 2) SDE (VE)
sde = VESDE(sigma_min=0.01, sigma_max=10.0, T=1.0)

# 3)  ScoreNet
score_net = ScoreNet(dim=2, hidden_dim=64).to(device)

# 4)  score network
train_score_network(score_net, sde, funnel_dist,
                    epochs=10000, batch_size=256, lr=1e-3, device=device)
```

```python
# 5) SDE
gen_samples = sample_reverse_sde(score_net, sde, num_samples=5000,
                                    num_steps=1500, device=device)
gen_samples = gen_samples.cpu().numpy()  # shape [5000, 2]

# 6) funnel
real_samples = funnel_dist.sample(5000).cpu().numpy()

fig, axs = plt.subplots(1,2, figsize=(10,4))

axs[0].scatter(real_samples[:,0], real_samples[:,1], alpha=0.3, s=5)
axs[0].set_title("Real Funnel Samples")
axs[0].set_xlabel("x"); axs[0].set_ylabel("y")

axs[1].scatter(gen_samples[:,0], gen_samples[:,1], alpha=0.3, s=5, color='orange')
axs[1].set_title("Generated via Reverse VE-SDE")
axs[1].set_xlabel("x"); axs[1].set_ylabel("y")

plt.tight_layout()
plt.show()
```
